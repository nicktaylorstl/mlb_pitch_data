```python
import pandas as pd
import requests
import time
import statsapi
from datetime import date
pd.set_option('display.max_columns', None)
```

```python
today = date.today()
today
```

```python
games = statsapi.schedule(start_date='2024-03-03',end_date=today)
G = pd.json_normalize(games)
R =G.loc[G['game_type']== 'R']
```

```python
R.head()
```

```python
import json
import csv
game_ids = sorted(R['game_id'].unique().astype(list))

existing_game_ids = set()
try:
    with open('game_info/game_IDs.csv', mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row:
                existing_game_ids.add(int(row[0]))
except FileNotFoundError:
    pass

count = 0
for gid in game_ids:
    count += 1
    if gid in existing_game_ids:
        print(f'Game {gid} previously recorded')
        continue

    time.sleep(5)
    url = 'https://statsapi.mlb.com/api/v1.1/game/' + str(gid)+'/feed/live'
    r= requests.get(url)
    j = json.loads(r.text)
    with open(f'game_info/{gid}.json', 'w') as f:
        json.dump(j, f, ensure_ascii=False)
    with open('game_info/game_IDs.csv', mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([gid])
    print(f"Game {gid} added to list ({count} of {len(game_ids)})")


```
