```python
# Recording Player Data
import csv
import json

existing_game_ids = set()
try:
    with open('game_data/game_IDs.csv', mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row:
                existing_game_ids.add(int(row[0]))
except FileNotFoundError:
    pass



game_ids = list(existing_game_ids)

# Loop through each game ID and process the corresponding JSON file
count = 1
for game_id in game_ids:
    print(f"Writing game {count} of {len(game_ids)}")
    count += 1

    existing_player_ids = set()
    try:
        with open('pitch_data/mlb_players.csv', mode='r') as file:
            reader = csv.reader(file)
            next(reader)  # Skip header row
            for row in reader:
                if row:
                    existing_player_ids.add(int(row[0]))
    except FileNotFoundError:
        pass

    file_path = f"game_data/{game_id}.json"
    with open(file_path, 'r') as file:
        data = json.load(file)

    players = data["gameData"]["players"]

    for player_id, player in players.items():
        player_id = player.get("id")

        if int(player_id) in existing_player_ids:
            print(f"{player_id} already recorded")
            continue  # Skip if player_id already exists

        uniform_number = player.get("primaryNumber")
        full_name = player.get("fullName")
        first_name= player.get("firstName")
        last_name = player.get("lastName")
        name_first_last = player.get("nameFirstLast")
        use_first_name= player.get("useName")
        use_last_name= player.get("useLastName")
        birth_date = player.get("birthDate")
        current_age = player.get("currentAge")
        birth_city = player.get("birthCity")
        birth_state_province= player.get("birthStateProvince")
        birth_country = player.get("birthCountry")
        height = player.get("height")
        if height:
            feet, inches = height.split("' ")
            feet = int(feet)
            inches = int(inches.replace('"', ''))
            height_inches = feet * 12 + inches
        else:
            height_inches = None

        weight = player.get("weight")
        position = player.get("primaryPosition", {}).get("code")

        mlb_debut = player.get("mlbDebutDate")
        bats = player.get("batSide",{}).get("code")
        throws = player.get("pitchHand",{}).get("code")
        



        row = player_id,uniform_number,full_name,first_name,last_name,use_first_name,use_last_name,name_first_last,birth_date,birth_city,birth_state_province,birth_country,current_age,height_inches,weight,position,mlb_debut,bats,throws
        with open('pitch_data/mlb_players.csv', mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(row)
        print(row)
```
